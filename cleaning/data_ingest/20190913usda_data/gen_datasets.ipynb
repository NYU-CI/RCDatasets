```python
def run_query(conn,query):
    cur = conn.cursor()
    cur.execute(query)
    rows = cur.fetchall()
    cur.close()
    return rows
```

```python
conn = db_connect()
query = "SELECT distinct canonical_name,data_provider FROM public.dataset_crosswalk;"
return_query = run_query(conn = conn, query = query)
```
